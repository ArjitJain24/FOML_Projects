```python
#Understanding Error Surface
#All imports
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import collections


########################################
#Generate meaningfull data
########################################
heights = np.random.uniform(48,73,25)
weights = np.array([3.86*h - 110.42 + np.random.normal(0,np.sqrt(20)) for h in heights])




########################################
#Plot scatter plot of data
########################################
plt.scatter(heights,weights, color='r')
plt.xlabel('Height(inches)')
plt.ylabel('Weight(pounds)')
plt.title('Scatter plot of Height vs Weight')
plt.show()


########################################
#Weight estimation through error surface, i.e., empirically locate the minima of error surface
########################################
#Sample a bunch of w's around w_opt and compute the associated error
X = np.concatenate((np.ones((25,1)),heights.reshape((25,1))),axis=1)
Xt_X = (X.T@ X)
Xt_X_inv = np.linalg.inv(Xt_X)
w_opt = ((Xt_X_inv@ X.T)@weights.reshape((25,1)))       # shape is (2,1)

#Sample w0 and w1
w0 = np.linspace(w_opt[0,0]-20,w_opt[0,0]+20,100)
w1 = np.linspace(w_opt[1,0]-1,w_opt[1,0]+1,100)

#Compute the error
def Error(w,t,x): #inputs : 1)weight 2)data i.e (t,x)
    #Estimate the target
    y_hat = np.array([w[0] + w[1]*sample for sample in x])
    #Compute and return the error
    error = (np.sum((t-y_hat)**2))/len(t)
    return error


#Plot 3D error surface and the corresponding contour plots
#Error surface is a function of w0 and w1
J = np.zeros((len(w0), len(w1)))
for i in range(len(w0)):
    for j in range(len(w1)):
        J[i, j] = Error([w0[i], w1[j]], weights, heights)

#Locate the minima of the error surface
index = np.argmin(J)
w0_optimum1 = w0[int(index/len(w1))]
w1_optimum1 = w1[int(index%len(w1))]
print("Optimal w0 from minimising Error surface: ", w0_optimum1)
print("Optimal w1 from minimising Error surface: ", w1_optimum1)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
W0, W1 = np.meshgrid(w0, w1)
ax.plot(w0_optimum1,w1_optimum1,'ro')
ax.plot_surface(W0, W1, J, cmap='viridis')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('Cost')
plt.show()

fig, ax = plt.subplots(1, 1)
ax.contour(W0, W1, J ,50)
ax.plot(w0_optimum1,w1_optimum1,'ro')
ax.set_title('Contour plot of the error surface')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
plt.show()




########################################
#Least squares approach to estimate the weights
########################################
#Complete the below linear regression function
def LinearRegression(x,t): #inputs : 1)input data i.e (x). 2)target i.e (t)
    X = np.concatenate((np.ones((len(x),1)),x.reshape((len(x),1))),axis=1)
    Xt_X = (X.T@ X)
    Xt_X_inv = np.linalg.inv(Xt_X)
    w_opt = ((Xt_X_inv@ X.T)@t.reshape((len(x),1)))
    return w_opt



#Estimate optimal weight's using "LinearRegression" function
w_opt = LinearRegression(heights,weights)
w0_optimum2 = w_opt[0,0]
w1_optimum2 = w_opt[1,0]
print("Optimal w0 from Linear Regression Function: ", w0_optimum2)
print("Optimal w1 from Linear Regression Function: ", w1_optimum2)

#Estimate the targets using the input x and the estimated weights
t_hat1 = np.array([w0_optimum1 + w1_optimum1*height for height in heights])
t_hat2 = np.array([w0_optimum2 + w1_optimum2*height for height in heights])


#Plot the estimated line on top of the above scatter plot
x = np.linspace(48,72,100)
plt.plot(x,w0_optimum2 + w1_optimum2*x, color='b')
plt.scatter(heights,weights, color='r')
plt.xlabel("w0")
plt.ylabel("w1")
plt.show()


########################################
#Compare the estimated weight's using least squares approach with the error surface approach
########################################
Error1 = Error([w0_optimum1,w1_optimum1],weights,heights)
Error2 = Error([w0_optimum2,w1_optimum2],weights,heights)
print("MSE from Error Surface Approach: ", Error1)
print("MSE from Least Squares Approach: ", Error2)


```

```python
#Understanding model order and overfitting
########################################
#Generate 20 data points
########################################
x = np.random.rand(20)
t = np.array([np.sin(2*np.pi*x) + np.random.normal(0,np.sqrt(0.1)) for x in x])


#Obtain train and test splits
x_train1 = x[0:10]
x_test1 = x[10:20]
t_train1 = t[0:10]
t_test1 = t[10:20]
########################################
#Fitting Mth degree polynomial using least squares approach
########################################

def PseudoInverse(phi,Y_train,lamda):
    M = len(phi[0,:])-1
    phi_t_phi = (phi.T@phi)
    temp = phi_t_phi + lamda*np.eye(M+1)
    phi_t_phi_inv = np.linalg.inv(temp)
    w_opt = ((phi_t_phi_inv@phi.T)@Y_train.reshape((len(phi[:,0]),1)))
    #return the weight vector
    return w_opt


#Complete the function
def PolynomialFit(X_train,Y_train,M,lamda): #(training data, training targets, Model order, Regularization coefficient)
    #Transform the data using polynomial kernel
    phi = np.zeros((len(X_train),M+1))
    for i in range(M+1):
        phi[:,i] = X_train**i
    return PseudoInverse(phi,Y_train,lamda)


#Complete the function
def PolynomialPred(w_est,X_train,X_test): #(weight,training data, testing data)
    #Estimate the targets for both training and testing data
    M = len(w_est)-1
    phi = np.zeros((len(X_train),M+1))
    for i in range(M+1):
        phi[:,i] = X_train**i
    TrainPred = (phi@w_est.reshape((M+1,1)))

    test_phi = np.zeros((len(X_test),M+1))
    for i in range(M+1):
        test_phi[:,i] = X_test**i
    TestPred = (test_phi@w_est.reshape((M+1,1)))

    TrainPred = TrainPred.reshape((1,len(X_train)))
    TestPred = TestPred.reshape((1,len(X_test)))
    #Return training and testing predictions
    return TrainPred,TestPred

#Complete the function
def PolynomialPred_Error(w_est,X_train,X_test,Y_train,Y_test): #(weight,training data, testing data, training targets, testing targets)
    #Estimate the targets for both training and testing data
    TrainPred,TestPred = PolynomialPred(w_est,X_train,X_test)
    #Compute the error for both training and testing data
    TrainError = (np.sum((Y_train-TrainPred)**2))/len(Y_train)
    TestError = (np.sum((Y_test-TestPred)**2))/len(Y_test)
    #Return training and testing error
    return TrainError,TestError

#Iterate through range of M values
M_range=list(range(10))
TrError = []
TeError = []
TrPred = []
TePred = []
lamda = 0
for M in M_range:
    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_opt = PolynomialFit(x_train1,t_train1,M,lamda)
    #Predict training and testing targets
    TrPred,TePred = PolynomialPred(w_opt,x_train1,x_test1)
    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    TrainError,TestError = PolynomialPred_Error(w_opt,x_train1,x_test1,t_train1,t_test1)

    #Store them for plotting
    TrError.append(TrainError)
    TeError.append(TestError)

    # Plot training and testing estimates alogwith the original targets
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle(f'M = {M}')
    ax1.set_title("Training Estimates vs Original Training Targets")
    ax1.scatter(x_train1,t_train1, color='r')
    ax1.scatter(x_train1,TrPred, color='g')
    ax1.set_xlabel('Training Data')
    ax1.set_ylabel('Training Targets')
    ax1.legend(["Original","Predicted"])
    ax2.set_title("Test Estimates vs Original Test Targets")
    ax2.scatter(x_test1,t_test1, color='r')
    ax2.scatter(x_test1,TePred, color='g')
    ax2.set_xlabel('Testing Data')
    ax2.set_ylabel('Testing Targets')
    ax2.legend(["Original","Predicted"])
    plt.show()


TrError = np.array(TrError)
TeError = np.array(TeError)


#Plot training error vs polynomial order, and testing error vs polynomial order
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,np.sqrt(TrError), color='r')
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training RMSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,np.sqrt(TeError), color='r')
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing RMSE')
plt.show()


########################################
#Increase the size of training data set to 100 points and repeat the experiments
########################################
x = np.random.rand(100)
t = np.array([np.sin(2*np.pi*x) + np.random.normal(0,np.sqrt(0.1)) for x in x])


#Obtain train and test splits
x_train2 = x[0:50]
x_test2 = x[50:100]
t_train2 = t[0:50]
t_test2 = t[50:100]



#Iterate through range of M values
M_range=list(range(10))
lamda = 0
TrError = []
TeError = []
TrPred = []
TePred = []
for M in M_range:
    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_opt = PolynomialFit(x_train2,t_train2,M,lamda)
    #Predict training and testing targets
    TrPred,TePred = PolynomialPred(w_opt,x_train2,x_test2)
    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    TrainError,TestError = PolynomialPred_Error(w_opt,x_train2,x_test2,t_train2,t_test2)

    #Store them for plotting
    TrError.append(TrainError)
    TeError.append(TestError)

    # Plot training and testing estimates alogwith the original targets
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle(f'M = {M}')
    ax1.set_title("Training Estimates vs Original Training Targets")
    ax1.scatter(x_train2,t_train2, color='r')
    ax1.scatter(x_train2,TrPred, color='g')
    ax1.set_xlabel('Training Data')
    ax1.set_ylabel('Training Targets')
    ax1.legend(["Original","Predicted"])
    ax2.set_title("Test Estimates vs Original Test Targets")
    ax2.scatter(x_test2,t_test2, color='r')
    ax2.scatter(x_test2,TePred, color='g')
    ax2.set_xlabel('Testing Data')
    ax2.set_ylabel('Testing Targets')
    ax2.legend(["Original","Predicted"])
    plt.show()


TrError = np.array(TrError)
TeError = np.array(TeError)


#Plot training error vs polynomial order, and testing error vs polynomial order
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,np.sqrt(TrError), color='r')
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training RMSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,np.sqrt(TeError), color='r')
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing RMSE')
plt.show()




########################################
#Effect of regularization
########################################
#Consider a set of lambda's. For example: lamdas = [0, 1e-7 , 1e-4, 1e-2, 1]
#Repeat the experiments, i.e., plot the prediction and error in predictions with respect to model order. Contrast these results with those obtained without regularization.
lamdas = [0 , 1e-7, 1e-4, 1e-2, 1]

TrError = []
TeError = []
TrPred = []
TePred = []
i = 0

for M in M_range:
    for lamda in lamdas:
        #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
        w_opt = PolynomialFit(x_train1,t_train1,M,lamda)
        #Predict training and testing targets
        TrainPred,TestPred = PolynomialPred(w_opt,x_train1,x_test1)
        #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
        TrainError,TestError = PolynomialPred_Error(w_opt,x_train1,x_test1,t_train1,t_test1)

        #Store them for plotting
        TrError.append(TrainError)
        TeError.append(TestError)
        TrPred.append(TrainPred)
        TePred.append(TestPred)
    # Plot training and testing estimates alogwith the original targets
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
    fig.suptitle(f'M = {M}')
    ax1.set_title("Training Estimates vs Original Training Targets")
    ax1.scatter(x_train1,t_train1, color='r', s = 30)
    ax1.scatter(x_train1,TrPred[i+0], color="black", s = 30)
    ax1.scatter(x_train1,TrPred[i+1], color="blue", s = 30)
    ax1.scatter(x_train1,TrPred[i+2], color="green", s = 30)
    ax1.scatter(x_train1,TrPred[i+3], color="orange", s = 30)
    ax1.scatter(x_train1,TrPred[i+4], color="purple", s = 30)
    ax1.set_xlabel('Training Data')
    ax1.set_ylabel('Training Targets')
    ax1.legend(["Original","Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
    ax2.set_title("Test Estimates vs Original Test Targets")
    ax2.scatter(x_test1,t_test1, color='r', s = 30)
    ax2.scatter(x_test1,TePred[i+0], color='black', s = 30)
    ax2.scatter(x_test1,TePred[i+1], color='blue', s = 30)
    ax2.scatter(x_test1,TePred[i+2], color='green', s = 30)
    ax2.scatter(x_test1,TePred[i+3], color='orange', s = 30)
    ax2.scatter(x_test1,TePred[i+4], color='purple', s = 30)
    ax2.set_xlabel('Testing Data')
    ax2.set_ylabel('Testing Targets')
    ax2.legend(["Original","Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
    plt.show()
    i+=len(lamdas)

TrError = np.array(TrError)
TeError = np.array(TeError)

#Plot training error vs polynomial order, and testing error vs polynomial order
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title("Training Error vs Polynomial Order")
for i in range(len(lamdas)):
    ax1.plot(M_range,np.sqrt(TrError[i::len(lamdas)]))
ax1.legend(["Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training RMSE')
ax2.set_title("Test Error vs Polynomial Order")
for i in range(len(lamdas)):
    ax2.plot(M_range,np.sqrt(TeError[i::len(lamdas)]))
ax2.legend(["Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing RMSE')
ax2.set_ylim(0, max(ax1.get_ylim()[1], ax2.get_ylim()[1]))
plt.show()





TrError = []
TeError = []
TrPred = []
TePred = []
i = 0
for M in M_range:
    for lamda in lamdas:
        #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
        w_opt = PolynomialFit(x_train2,t_train2,M,lamda)
        #Predict training and testing targets
        TrainPred,TestPred = PolynomialPred(w_opt,x_train2,x_test2)
        #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
        TrainError,TestError = PolynomialPred_Error(w_opt,x_train2,x_test2,t_train2,t_test2)

        #Store them for plotting
        TrError.append(TrainError)
        TeError.append(TestError)
        TrPred.append(TrainPred)
        TePred.append(TestPred)
    # Plot training and testing estimates alogwith the original targets
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
    fig.suptitle(f'M = {M}')
    ax1.set_title("Training Estimates vs Original Training Targets")
    ax1.scatter(x_train2,t_train2, color='r', s = 30)
    ax1.scatter(x_train2,TrPred[i+0], color="black", s = 30)
    ax1.scatter(x_train2,TrPred[i+1], color="blue", s = 30)
    ax1.scatter(x_train2,TrPred[i+2], color="green", s = 30)
    ax1.scatter(x_train2,TrPred[i+3], color="orange", s = 30)
    ax1.scatter(x_train2,TrPred[i+4], color="purple", s = 30)
    ax1.set_xlabel('Training Data')
    ax1.set_ylabel('Training Targets')
    ax1.legend(["Original","Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
    ax2.set_title("Test Estimates vs Original Test Targets")
    ax2.scatter(x_test2,t_test2, color='r', s = 30)
    ax2.scatter(x_test2,TePred[i+0], color='black', s = 30)
    ax2.scatter(x_test2,TePred[i+1], color='blue', s = 30)
    ax2.scatter(x_test2,TePred[i+2], color='green', s = 30)
    ax2.scatter(x_test2,TePred[i+3], color='orange', s = 30)
    ax2.scatter(x_test2,TePred[i+4], color='purple', s = 30)
    ax2.set_xlabel('Testing Data')
    ax2.set_ylabel('Testing Targets')
    ax2.legend(["Original","Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
    plt.show()
    i+=len(lamdas)
TrError = np.array(TrError)
TeError = np.array(TeError)
# #Plot training error vs polynomial order, and testing error vs polynomial order
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title("Training Error vs Polynomial Order")
for i in range(len(lamdas)):
    ax1.plot(M_range,np.sqrt(TrError[i::len(lamdas)]))
ax1.legend(["Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training RMSE')
ax2.set_title("Test Error vs Polynomial Order")
for i in range(len(lamdas)):
    ax2.plot(M_range,np.sqrt(TeError[i::len(lamdas)]))
ax2.legend(["Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing RMSE')
ax2.set_ylim(0, max(ax1.get_ylim()[1], ax2.get_ylim()[1]))
plt.show()





########################################
#Effect of bias regularization
########################################
#Modify the function i.e include bias
#Generate data
x = np.random.rand(100)
t = np.array([np.sin(2*np.pi*x) + 5 + np.random.normal(0,np.sqrt(0.1)) for x in x])

x_train = x[0:50]
x_test = x[50:100]
t_train = t[0:50]
t_test = t[50:100]

#Estimate the polynomial with and without regularization constraint
lamda = 0
TrError_noLam = []
TeError_noLam = []
TrPred_noLam = []
TePred_noLam = []
for M in M_range:
    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_opt = PolynomialFit(x_train,t_train,M,0)
    #Predict training and testing targets
    TrPred_noLam,TePred_noLam = PolynomialPred(w_opt,x_train,x_test)
    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    TrainError,TestError = PolynomialPred_Error(w_opt,x_train,x_test,t_train,t_test)

    #Store them for plotting
    TrError_noLam.append(TrainError)
    TeError_noLam.append(TestError)

    # Plot training and testing estimates alogwith the original targets
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle(f'M = {M}')
    ax1.set_title("Training Estimates vs Original Training Targets")
    ax1.scatter(x_train,t_train, color='r')
    ax1.scatter(x_train,TrPred_noLam, color='g')
    ax1.set_xlabel('Training Data')
    ax1.set_ylabel('Training Targets')
    ax1.legend(["Original","Predicted"])
    ax2.set_title("Test Estimates vs Original Test Targets")
    ax2.scatter(x_test,t_test, color='r')
    ax2.scatter(x_test,TePred_noLam, color='g')
    ax2.set_xlabel('Testing Data')
    ax2.set_ylabel('Testing Targets')
    ax2.legend(["Original","Predicted"])
    plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,np.sqrt(TrError_noLam), color='r')
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training RMSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,np.sqrt(TeError_noLam), color='r')
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing RMSE')
ax2.set_ylim(0, max(ax1.get_ylim()[1], ax2.get_ylim()[1]))
plt.show()

TrError_Lam = []
TeError_Lam = []
TrPred_Lam = []
TePred_Lam = []
i = 0
for M in M_range:
    for lamda in lamdas:
        #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
        w_opt = PolynomialFit(x_train,t_train,M,lamda)
        #Predict training and testing targets
        TrainPred,TestPred = PolynomialPred(w_opt,x_train,x_test)
        #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
        TrainError,TestError = PolynomialPred_Error(w_opt,x_train,x_test,t_train,t_test)

        #Store them for plotting
        TrError_Lam.append(TrainError)
        TeError_Lam.append(TestError)
        TrPred_Lam.append(TrainPred)
        TePred_Lam.append(TestPred)



    # Plot training and testing estimates alogwith the original targets
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
    fig.suptitle(f'M = {M}')
    ax1.set_title("Training Estimates vs Original Training Targets")
    ax1.scatter(x_train,t_train, color='r', s = 30)
    ax1.scatter(x_train,TrPred_Lam[i+0], color="black", s = 30)
    ax1.scatter(x_train,TrPred_Lam[i+1], color="blue", s = 30)
    ax1.scatter(x_train,TrPred_Lam[i+2], color="green", s = 30)
    ax1.scatter(x_train,TrPred_Lam[i+3], color="orange", s = 30)
    ax1.scatter(x_train,TrPred_Lam[i+4], color="purple", s = 30)
    ax1.set_xlabel('Training Data')
    ax1.set_ylabel('Training Targets')
    ax1.legend(["Original","Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
    ax2.set_title("Test Estimates vs Original Test Targets")
    ax2.scatter(x_test,t_test, color='r', s = 30)
    ax2.scatter(x_test,TePred_Lam[i+0], color='black', s = 30)
    ax2.scatter(x_test,TePred_Lam[i+1], color='blue', s = 30)
    ax2.scatter(x_test,TePred_Lam[i+2], color='green', s = 30)
    ax2.scatter(x_test,TePred_Lam[i+3], color='orange', s = 30)
    ax2.scatter(x_test,TePred_Lam[i+4], color='purple', s = 30)
    ax2.set_xlabel('Testing Data')
    ax2.set_ylabel('Testing Targets')
    ax2.legend(["Original","Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
    plt.show()
    i+=len(lamdas)


TrError_Lam = np.array(TrError_Lam)
TeError_Lam = np.array(TeError_Lam)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title("Training Error vs Polynomial Order")
for i in range(len(lamdas)):
    ax1.plot(M_range,np.sqrt(TrError_Lam[i::len(lamdas)]))
ax1.legend(["Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training RMSE')
ax2.set_title("Test Error vs Polynomial Order")
for i in range(len(lamdas)):
    ax2.plot(M_range,np.sqrt(TeError_Lam[i::len(lamdas)]))
ax2.legend(["Lamda = 0", "Lamda = 1e-7", "Lamda = 1e-4", "Lamda = 1e-2", "Lamda = 1"])
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing RMSE')
ax2.set_ylim(0, max(ax1.get_ylim()[1], ax2.get_ylim()[1]))
plt.show()
#Compare the two estimated polynomials and report the observations

```

```python
#Understanding the choice of kernel
########################################
#Generate 100 data points
########################################

x = np.random.rand(100)
t = np.array([np.sin(2*np.pi*x) + np.random.normal(0,np.sqrt(0.1)) for x in x])

#Obtain train and test splits
#Take even samples for training and odd samples for testing
x_train_1 = x[1:100:2]
x_test_1 = x[0:100:2]
t_train_1 = t[1:100:2]
t_test_1 = t[0:100:2]


#Function to estimate the parameters
def KernelRegressionFit(X_train,Y_train,kernelType,M,lamda): #(training data, training targets, type of kernel, regularization coefficient)
    #kernelType : {Polynomial,Gaussian,Sigmoid}
    #print(X_train.shape)
    #X_train = np.reshape(X_train, (1,-1))
    #Y_train = np.reshape(Y_train, (1,-1))
    X = np.zeros((len(X_train),M+1))
    if kernelType=='polynomial':
      #Use polynomial kernel to transform the data
      for i in range(M+1):
        X[:,i] = X_train**i

    if kernelType=='gaussian':
      #Use Gaussian kernel to transform the data
      u = np.linspace(np.min(X_train),np.max(X_train),M+1)
      for i in range(M+1):
        X[:,i] = np.exp(-0.5*((X_train-u[i])**2))

    if kernelType=='sigmoidal':
      #Use Sigmoid kernel to transform the data
      u = np.linspace(np.min(X_train),np.max(X_train),M+1)
      for i in range(M+1):
        X[:,i] = np.tanh((X_train-u[i])+0.1)

    #Estimate weights using Pseudo iverse solution
    w_opt = PseudoInverse(X,Y_train,lamda)

    #Return the estimated weights
    return w_opt

#Function to compute the training and testing errors from the current weight estimates
def KernelRegressionPred_Error(w_est,X_train,Y_train,X_test,Y_test,kernelType):
    #kernelType : {Polynomial,Gaussian,Sigmoid}
    X_tr = []
    X_te = []
    #X_train = np.reshape(X_train, (1,-1))
    #Y_train = np.reshape(Y_train, (1,-1))
    #X_test = np.reshape(X_test, (1,-1))
    #Y_test = np.reshape(Y_test, (1,-1))
    M = len(w_est)-1
    X_tr = np.zeros((len(X_train),M+1))
    X_te = np.zeros((len(X_test),M+1))
    if kernelType=='polynomial':
      #Use polynomial kernel to transform the data
      for i in range(M+1):
        X_tr[:,i] = X_train**i
        X_te[:,i] = X_test**i

    if kernelType=='gaussian':
      #Use Gaussian kernel to transform the data
      u_tr = np.linspace(np.min(X_train),np.max(X_train),M+1)
      u_te = np.linspace(np.min(X_test),np.max(X_test),M+1)
      for i in range(M+1):
        X_tr[:,i] = np.exp(-0.5*((X_train-u_tr[i])**2))
        X_te[:,i] = np.exp(-0.5*((X_test-u_te[i])**2))


    if kernelType=='sigmoidal':
      #Use Sigmoid kernel to transform the data
      u_tr = np.linspace(np.min(X_train),np.max(X_train),M+1)
      u_te = np.linspace(np.min(X_test),np.max(X_test),M+1)
      for i in range(M+1):
        X_tr[:,i] = np.tanh((X_train-u_tr[i])+0.1)
        X_te[:,i] = np.tanh((X_test-u_te[i])+0.1)



    #Estimate training and testing targets
    TrainPred = (X_tr@w_est.reshape((M+1,1)))
    TestPred = (X_te@w_est.reshape((M+1,1)))

    #Compute and return the training and testing errors
    TrainError = (np.sum((Y_train-TrainPred)**2))/len(Y_train)
    TestError = (np.sum((Y_test-TestPred)**2))/len(Y_test)
    return TrainPred, TestPred, TrainError,TestError


#Iterate through range of M values
M_range=list(range(10))

polynomial_tr_error = []
polynomial_te_error = []
gaussian_tr_error = []
gaussian_te_error = []
sigmoid_tr_error = []
sigmoid_te_error = []
lamda = 0

for M in M_range:
    #Fit Mth order polynomial using three kernels i.e {Polynomial,Gaussian,Sigmoid}
    w_opt_po = KernelRegressionFit(x_train_1,t_train_1,"polynomial",M,lamda)
    w_opt_ga = KernelRegressionFit(x_train_1,t_train_1,"gaussian",M,lamda)
    w_opt_si = KernelRegressionFit(x_train_1,t_train_1,"sigmoidal",M,lamda)

# TrError = np.array(TrError)
# TeError = np.array(TeError)


    #Predict training and testing targets using estimated w
    #Predict errors on both training and testing data using estimated w
    TrainPred_po , TestPred_po , TrainError_po, TestError_po = KernelRegressionPred_Error(w_opt_po,x_train_1,t_train_1,x_test_1,t_test_1,"polynomial")
    TrainPred_ga , TestPred_ga , TrainError_ga, TestError_ga = KernelRegressionPred_Error(w_opt_ga,x_train_1,t_train_1,x_test_1,t_test_1,"gaussian")
    TrainPred_si , TestPred_si , TrainError_si, TestError_si = KernelRegressionPred_Error(w_opt_si,x_train_1,t_train_1,x_test_1,t_test_1,"sigmoidal")



    #Plot the predicted training and testing targets alongside the original targets for various model orders and all three different kernels.
    fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(15, 15))
    fig.suptitle(f'M = {M}', fontweight='bold')

    fig.text(0.5, 0.96, 'Polynomial Kernel ', ha='center', fontsize=10, fontweight='bold', color = 'red')
    ax1.set_title("Training Estimates vs Original Training Targets")
    ax1.scatter(x_train_1,t_train_1, color='r')
    ax1.scatter(x_train_1,TrainPred_po, color='g')
    ax1.set_xlabel('Training Data')
    ax1.set_ylabel('Training Targets')
    ax1.legend(["Original","Predicted"])
    ax2.set_title("Test Estimates vs Original Test Targets")
    ax2.scatter(x_test_1,t_test_1, color='r')
    ax2.scatter(x_test_1,TestPred_po, color='g')
    ax2.set_xlabel('Testing Data')
    ax2.set_ylabel('Testing Targets')
    ax2.legend(["Original","Predicted"])

    fig.text(0.5, 0.64, 'Gaussian Kernel ', ha='center', fontsize=10, fontweight='bold', color = 'red')
    ax3.set_title("Training Estimates vs Original Training Targets")
    ax3.scatter(x_train_1,t_train_1, color='r')
    ax3.scatter(x_train_1,TrainPred_ga, color='g')
    ax3.set_xlabel('Training Data')
    ax3.set_ylabel('Training Targets')
    ax3.legend(["Original","Predicted"])
    ax4.set_title("Test Estimates vs Original Test Targets")
    ax4.scatter(x_test_1,t_test_1, color='r')
    ax4.scatter(x_test_1,TestPred_ga, color='g')
    ax4.set_xlabel('Testing Data')
    ax4.set_ylabel('Testing Targets')
    ax4.legend(["Original","Predicted"])

    fig.text(0.5, 0.32, 'Sigmoidal Kernel ', ha='center', fontsize=10, fontweight='bold', color = 'red')
    ax5.set_title("Training Estimates vs Original Training Targets")
    ax5.scatter(x_train_1,t_train_1, color='r')
    ax5.scatter(x_train_1,TrainPred_si, color='g')
    ax5.set_xlabel('Training Data')
    ax5.set_ylabel('Training Targets')
    ax5.legend(["Original","Predicted"])
    ax6.set_title("Test Estimates vs Original Test Targets")
    ax6.scatter(x_test_1,t_test_1, color='r')
    ax6.scatter(x_test_1,TestPred_si, color='g')
    ax6.set_xlabel('Testing Data')
    ax6.set_ylabel('Testing Targets')
    ax6.legend(["Original","Predicted"])

    plt.tight_layout()
    plt.show()

    #Store them for plotting
    polynomial_tr_error.append(TrainError_po)
    polynomial_te_error.append(TestError_po)
    gaussian_tr_error.append(TrainError_ga)
    gaussian_te_error.append(TestError_ga)
    sigmoid_tr_error.append(TrainError_si)
    sigmoid_te_error.append(TestError_si)

polynomial_tr_error = np.array(polynomial_tr_error)
polynomial_te_error = np.array(polynomial_te_error)
gaussian_tr_error = np.array(gaussian_tr_error)
gaussian_te_error = np.array(gaussian_te_error)
sigmoid_tr_error = np.array(sigmoid_tr_error)
sigmoid_te_error = np.array(sigmoid_te_error)



#Plot training error vs polynomial order and testing error vs polynomial order for all the three different kernels

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('Training and Testing Errors vs Polynomial Order', fontweight='bold')

fig.text(0.5, 0.96, 'Polynomial Kernel ', ha='center', fontsize=10, fontweight='bold', color = 'red')
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,np.sqrt(polynomial_tr_error), color='r')
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training RMSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,np.sqrt(polynomial_te_error), color='r')
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing RMSE')

fig.text(0.5, 0.64, 'Gaussian Kernel ', ha='center', fontsize=10, fontweight='bold', color = 'red')
ax3.set_title("Training Error vs Polynomial Order")
ax3.plot(M_range,np.sqrt(gaussian_tr_error), color='r')
ax3.set_xlabel('Polynomial Order')
ax3.set_ylabel('Training RMSE')
ax4.set_title("Test Error vs Polynomial Order")
ax4.plot(M_range,np.sqrt(gaussian_te_error), color='r')
ax4.set_xlabel('Polynomial Order')
ax4.set_ylabel('Testing RMSE')

fig.text(0.5, 0.32, 'Sigmoidal Kernel ', ha='center', fontsize=10, fontweight='bold', color = 'red')
ax5.set_title("Training Error vs Polynomial Order")
ax5.plot(M_range,np.sqrt(sigmoid_tr_error), color='r')
ax5.set_xlabel('Polynomial Order')
ax5.set_ylabel('Training RMSE')
ax6.set_title("Test Error vs Polynomial Order")
ax6.plot(M_range,np.sqrt(sigmoid_te_error), color='r')
ax6.set_xlabel('Polynomial Order')
ax6.set_ylabel('Testing RMSE')

plt.tight_layout()
plt.show()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10))
fig.suptitle('Training and Testing Errors vs Polynomial Order', fontweight='bold')
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,np.sqrt(polynomial_tr_error), color='red')
ax1.plot(M_range,np.sqrt(gaussian_tr_error), color='blue')
ax1.plot(M_range,np.sqrt(sigmoid_tr_error), color='green')
ax1.legend(["Polynomial","Gaussian","Sigmoid"])
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training RMSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,np.sqrt(polynomial_te_error), color='red')
ax2.plot(M_range,np.sqrt(gaussian_te_error), color='blue')
ax2.plot(M_range,np.sqrt(sigmoid_te_error), color='green')
ax2.legend(["Polynomial","Gaussian","Sigmoid"])
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing RMSE')

plt.tight_layout()
plt.show()


########################################
#Repeat the experiments by changing target function
########################################

x1 = np.linspace(0,1,100)
np.random.shuffle(x1)
x2 = np.linspace(1,2,100)
np.random.shuffle(x2)
x3 = np.linspace(2,3,100)
np.random.shuffle(x3)

t1 = np.array([np.sin(2*np.pi*x) + np.random.normal(0,np.sqrt(0.1)) for x in x1])
t2= np.array([0.75 - np.absolute(x-1.75) + np.random.normal(0,np.sqrt(0.1)) for x in x2])
t3 = np.array([np.exp( -((x-2.5)**2)/0.3) + np.random.normal(0,np.sqrt(0.1)) for x in x3])

x1_train = x1[1:100:2]
x1_test = x1[0:100:2]
t1_train = t1[1:100:2]
t1_test = t1[0:100:2]
x2_train = x2[1:100:2]
x2_test = x2[0:100:2]
t2_train = t2[1:100:2]
t2_test = t2[0:100:2]
x3_train = x3[1:100:2]
x3_test = x3[0:100:2]
t3_train = t3[1:100:2]
t3_test = t3[0:100:2]
x_train_2 = np.concatenate((x1_train,x2_train,x3_train))
x_test_2 = np.concatenate((x1_test,x2_test,x3_test))
t_train_2 = np.concatenate((t1_train,t2_train,t3_train))
t_test_2 = np.concatenate((t1_test,t2_test,t3_test))

lamda = np.exp(-18)

#Iterate through range of M values
M_range=list(range(10))

polynomial_tr_error = []
polynomial_te_error = []
gaussian_tr_error = []
gaussian_te_error = []
sigmoid_tr_error = []
sigmoid_te_error = []

for M in M_range:
    #Fit Mth order polynomial using three kernels i.e {Polynomial,Gaussian,Sigmoid}
    w_opt_po = KernelRegressionFit(x_train_2,t_train_2,"polynomial",M,lamda)
    w_opt_ga = KernelRegressionFit(x_train_2,t_train_2,"gaussian",M,lamda)
    w_opt_si = KernelRegressionFit(x_train_2,t_train_2,"sigmoidal",M,lamda)


    #Predict training and testing targets using estimated w
    #Predict errors on both training and testing data using estimated w
    TrainPred_po , TestPred_po , TrainError_po, TestError_po = KernelRegressionPred_Error(w_opt_po,x_train_2,t_train_2,x_test_2,t_test_2,"polynomial")
    TrainPred_ga , TestPred_ga , TrainError_ga, TestError_ga = KernelRegressionPred_Error(w_opt_ga,x_train_2,t_train_2,x_test_2,t_test_2,"gaussian")
    TrainPred_si , TestPred_si , TrainError_si, TestError_si = KernelRegressionPred_Error(w_opt_si,x_train_2,t_train_2,x_test_2,t_test_2,"sigmoidal")



    #Plot the predicted training and testing targets alongside the original targets for various model orders and all three different kernels.
    fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(15, 15))
    fig.suptitle(f'M = {M}', fontweight='bold')

    fig.text(0.5, 0.96, 'Polynomial Kernel ', ha='center', fontsize=10, fontweight='bold')
    ax1.set_title("Training Estimates vs Original Training Targets")
    ax1.scatter(x_train_2,t_train_2, color='r')
    ax1.scatter(x_train_2,TrainPred_po, color='g')
    ax1.set_xlabel('Training Data')
    ax1.set_ylabel('Training Targets')
    ax1.legend(["Original","Predicted"])
    ax2.set_title("Test Estimates vs Original Test Targets")
    ax2.scatter(x_test_2,t_test_2, color='r')
    ax2.scatter(x_test_2,TestPred_po, color='g')
    ax2.set_xlabel('Testing Data')
    ax2.set_ylabel('Testing Targets')
    ax2.legend(["Original","Predicted"])

    fig.text(0.5, 0.64, 'Gaussian Kernel ', ha='center', fontsize=10, fontweight='bold')
    ax3.set_title("Training Estimates vs Original Training Targets")
    ax3.scatter(x_train_2,t_train_2, color='r')
    ax3.scatter(x_train_2,TrainPred_ga, color='g')
    ax3.set_xlabel('Training Data')
    ax3.set_ylabel('Training Targets')
    ax3.legend(["Original","Predicted"])
    ax4.set_title("Test Estimates vs Original Test Targets")
    ax4.scatter(x_test_2,t_test_2, color='r')
    ax4.scatter(x_test_2,TestPred_ga, color='g')
    ax4.set_xlabel('Testing Data')
    ax4.set_ylabel('Testing Targets')
    ax4.legend(["Original","Predicted"])

    fig.text(0.5, 0.32, 'Sigmoidal Kernel ', ha='center', fontsize=10, fontweight='bold')
    ax5.set_title("Training Estimates vs Original Training Targets")
    ax5.scatter(x_train_2,t_train_2, color='r')
    ax5.scatter(x_train_2,TrainPred_si, color='g')
    ax5.set_xlabel('Training Data')
    ax5.set_ylabel('Training Targets')
    ax5.legend(["Original","Predicted"])
    ax6.set_title("Test Estimates vs Original Test Targets")
    ax6.scatter(x_test_2,t_test_2, color='r')
    ax6.scatter(x_test_2,TestPred_si, color='g')
    ax6.set_xlabel('Testing Data')
    ax6.set_ylabel('Testing Targets')
    ax6.legend(["Original","Predicted"])

    plt.tight_layout()
    plt.show()

    #Store them for plotting
    polynomial_tr_error.append(TrainError_po)
    polynomial_te_error.append(TestError_po)
    gaussian_tr_error.append(TrainError_ga)
    gaussian_te_error.append(TestError_ga)
    sigmoid_tr_error.append(TrainError_si)
    sigmoid_te_error.append(TestError_si)

polynomial_tr_error = np.array(polynomial_tr_error)
polynomial_te_error = np.array(polynomial_te_error)
gaussian_tr_error = np.array(gaussian_tr_error)
gaussian_te_error = np.array(gaussian_te_error)
sigmoid_tr_error = np.array(sigmoid_tr_error)
sigmoid_te_error = np.array(sigmoid_te_error)






#Plot training error vs polynomial order and testing error vs polynomial order for all the three different kernels

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('Training and Testing Errors vs Polynomial Order', fontweight='bold')

fig.text(0.5, 0.96, 'Polynomial Kernel ', ha='center', fontsize=10, fontweight='bold')
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,np.sqrt(polynomial_tr_error), color='r')
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training RMSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,np.sqrt(polynomial_te_error), color='r')
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing RMSE')

fig.text(0.5, 0.64, 'Gaussian Kernel ', ha='center', fontsize=10, fontweight='bold')
ax3.set_title("Training Error vs Polynomial Order")
ax3.plot(M_range,np.sqrt(gaussian_tr_error), color='r')
ax3.set_xlabel('Polynomial Order')
ax3.set_ylabel('Training RMSE')
ax4.set_title("Test Error vs Polynomial Order")
ax4.plot(M_range,np.sqrt(gaussian_te_error), color='r')
ax4.set_xlabel('Polynomial Order')
ax4.set_ylabel('Testing RMSE')

fig.text(0.5, 0.32, 'Sigmoidal Kernel ', ha='center', fontsize=10, fontweight='bold')
ax5.set_title("Training Error vs Polynomial Order")
ax5.plot(M_range,np.sqrt(sigmoid_tr_error), color='r')
ax5.set_xlabel('Polynomial Order')
ax5.set_ylabel('Training RMSE')
ax6.set_title("Test Error vs Polynomial Order")
ax6.plot(M_range,sigmoid_te_error, color='r')
ax6.set_xlabel('Polynomial Order')
ax6.set_ylabel('Testing RMSE')

plt.tight_layout()
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10))
fig.suptitle('Training and Testing Errors vs Polynomial Order', fontweight='bold')
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,np.sqrt(polynomial_tr_error), color='red')
ax1.plot(M_range,np.sqrt(gaussian_tr_error), color='blue')
ax1.plot(M_range,np.sqrt(sigmoid_tr_error), color='green')
ax1.legend(["Polynomial","Gaussian","Sigmoid"])
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training RMSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,np.sqrt(polynomial_te_error), color='red')
ax2.plot(M_range,np.sqrt(gaussian_te_error), color='blue')
ax2.plot(M_range,np.sqrt(sigmoid_te_error), color='green')
ax2.legend(["Polynomial","Gaussian","Sigmoid"])
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing RMSE')

plt.tight_layout()
plt.show()






```

```python
#Understanding online training
########################################
#Repeat 3(a) and 3(b) using stochastic gradient descent for weight update ( plot required results )
########################################

def ErrorPred(w_est,X_train,Y_train,X_test,Y_test,kernelType): #(estimated weight, training data, training targets, testing data, testing targets, type of the kernel )
    #kernelType : {Polynomial,Gaussian,Sigmoid}
    M = len(w_est)-1
    X_tr = np.zeros((len(X_train),M+1))
    X_te = np.zeros((len(X_test),M+1))
    if kernelType=='polynomial':
        #Use polynomial kernel to transform the data
        for i in range(M+1):
            X_tr[:,i] = X_train**i
            X_te[:,i] = X_test**i

    if kernelType=='gaussian':
        #Use Gaussian kernel to transform the data
        u = np.linspace(np.min(X_train),np.max(X_train),M+1)
        for i in range(M+1):
            X_tr[:,i] = np.exp(-0.5*((X_train-u[i])**2))
            X_te[:,i] = np.exp(-0.5*((X_test-u[i])**2))

    if kernelType=='sigmoidal':
        #Use Sigmoid kernel to transform the data
        u_tr = np.linspace(np.min(X_train),np.max(X_train),M+1)
        u_te = np.linspace(np.min(X_test),np.max(X_test),M+1)
        for i in range(M+1):
            X_tr[:,i] = np.tanh((X_train-u_tr[i]) + 0.1)
            X_te[:,i] = np.tanh((X_test-u_te[i]) + 0.1)


    #Estimate training and testing targets
    TrainPred = (X_tr@w_est.reshape((M+1,1)))
    TestPred = (X_te@w_est.reshape((M+1,1)))

    #Compute and return the training and testing errors
    TrainError = (np.sum((Y_train-TrainPred)**2))/len(Y_train)
    TestError =  (np.sum((Y_test-TestPred)**2))/len(Y_test)
    return TrainError,TestError


def OnlineTraining(X_train,Y_train,X_test, Y_test, kernelType,M,Epochs,BatchSize,stepSize): #(training data, training targets, testing data, testing targets, tupe of the kernel, order of the mode, Number of epochs, Batch size, Step size)
    #kernelType : {Polynomial,Gaussian,Sigmoid}
    # Each time a dataset passes through an algorithm, it is said to have completed an epoch. Therefore, Epoch, in machine learning, refers to the one entire passing of training data through the algorithm.
    #Initialize the weights
    # Iterations is the number of batches needed to complete one epoch.
    weights = np.array([i for i in range(M+1)])
    #Initialize the necessary variables
    weights = weights.reshape((M+1,1))
    temp = weights
    sum_grad = 0
    #Iterate through epochs
    TrainError = []
    TestError = []
    grads = []
    length = 0
    epochs = range(Epochs)
    for epoch in epochs:
        # #Compute the train and test errors using the current weights
        tr_err, te_err = ErrorPred(weights, X_train, Y_train, X_test, Y_test, kernelType)
        #Store training and testing errors for plotting
        TrainError.append(tr_err)
        TestError.append(te_err)
        #Shuffle the data
        shuffled = np.random.permutation(len(X_train))
        X_train = X_train[shuffled]
        Y_train = Y_train[shuffled]
        #Iterate through the batches
        batches = int(len(X_train)/BatchSize)
        for batch in range(batches):
            #Initialize the necessary variables
            #Get a batch of data
            data_b = X_train[batch*BatchSize:(batch+1)*BatchSize]
            #Iterate through the data points of obtained batch
            for n in range(len(data_b)):
                #Obtain kernel representation
                X_tr = np.zeros((M+1,1))
                if kernelType=='polynomial':
                    #Use polynomial kernel to transform the data
                    for i in range(M+1):
                        X_tr[i] = data_b[n]**i

                if kernelType=='gaussian':
                    #Use Gaussian kernel to transform the data
                    u = np.linspace(np.min(data_b),np.max(data_b),M+1)
                    for i in range(M+1):
                        X_tr[i] = np.exp(-0.5*((data_b[n]-u[i])**2))

                if kernelType=='sigmoidal':
                    #Use Sigmoid kernel to transform the data
                    u = np.linspace(np.min(data_b),np.max(data_b),M+1)
                    for i in range(M+1):
                        X_tr[i] = np.tanh((data_b[n]-u[i])+0.1)

                #Compute the gradient of weight's
                grad = (Y_train[n + batch*BatchSize] - np.dot(X_tr.reshape((1,M+1)), weights))*X_tr
                # temp = temp + stepSize*(Y_train[n + batch*BatchSize] - np.dot(X_tr.reshape((1,M+1)), temp))*X_tr
                #Compute the running mean of the weights gradients for the batch update
                sum_grad = sum_grad + grad
                length += 1
                mean = sum_grad/length
                #Update the weights using mean gradient, consider using reasonable stepSize
                weights = weights + stepSize*mean

        # weights = weights + stepSize*mean_grad
       #Compute the train and test errors using the current weights
        # tr_err, te_err = ErrorPred(weights, X_train, Y_train, X_test, Y_test, kernelType)

       # print(tr_err,te_err)
        # TrainError.append(tr_err)
        # TestError.append(te_err)
        #Store training and testing errors for plotting

     #   gradients=[]
        #Shuffle the data

       # print(TrainError,TestError)
    #Plot training and testing error across the epochs
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    fig.suptitle(f'Mean Square Error vs Epochs : Kernel = {kernelType} : M = {M}', fontweight='bold')
    ax1.set_title("Training Error vs Epochs")
    ax1.set_xlim(0,Epochs+1,1)
    ax1.plot(range(1,Epochs+1,1),np.sqrt(TrainError), color='r')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Training Errors')
    ax2.set_title("Test Error vs Epochs")
    ax2.set_xlim(0,Epochs+1,1)
    ax2.plot(range(1,Epochs+1,1),np.sqrt(TestError), color='r')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Testing Errors')
    plt.show()

    #Return the estimated weights
    return weights




def OnlinePred(w_est,X_train,X_test,kernelType): #(estimated weights, training data, testing data, type of the kernel )
    #kernelType : {Polynomial,Gaussian,Sigmoid}
    #Initialize the required variables

    X_tr = []
    X_te = []
    M = len(w_est)-1
    X_tr = np.zeros((len(X_train),M+1))
    X_te = np.zeros((len(X_test),M+1))
    if kernelType=='polynomial':
      #Use polynomial kernel to transform the data
      for i in range(M+1):
        X_tr[:,i] = X_train**i
        X_te[:,i] = X_test**i

    if kernelType=='gaussian':
      #Use Gaussian kernel to transform the data
      u_tr = np.linspace(np.min(X_train),np.max(X_train),M+1)
      u_te = np.linspace(np.min(X_test),np.max(X_test),M+1)
      for i in range(M+1):
        X_tr[:,i] = np.exp(-0.5*((X_train-u_tr[i])**2))
        X_te[:,i] = np.exp(-0.5*((X_test-u_te[i])**2))


    if kernelType=='sigmoidal':
      #Use Sigmoid kernel to transform the data
      u_tr = np.linspace(np.min(X_train),np.max(X_train),M+1)
      u_te = np.linspace(np.min(X_test),np.max(X_test),M+1)
      for i in range(M+1):
        X_tr[:,i] = np.tanh((X_train-u_tr[i]) + 0.1)
        X_te[:,i] = np.tanh((X_test-u_te[i]) + 0.1)


    #Estimate training and testing targets
    TrainPred = (X_tr@w_est.reshape((M+1,1)))
    TestPred = (X_te@w_est.reshape((M+1,1)))

    return TrainPred,TestPred

def OnlinePred_Error(w_est,X_train,Y_train,X_test,Y_test,kernelType): #(estimated weights, training data, training targets, testing data, testing targets, type of the kernel )
    #kernelType : {Polynomial,Gaussian,Sigmoid}
    #Initialize the required variables

    #Obtain kernel representations
    if kernelType=='polynomial':
        TrainPred, TestPred = OnlinePred(w_est,X_train,X_test,kernelType)
    if kernelType=='gaussian':
        TrainPred, TestPred = OnlinePred(w_est,X_train,X_test,kernelType)
    if kernelType=='sigmoidal':
        TrainPred, TestPred = OnlinePred(w_est,X_train,X_test,kernelType)
    #Compute and return the training and testing errors
    TrainError = (np.sum((Y_train-TrainPred)**2))/len(Y_train)
    TestError = (np.sum((Y_test-TestPred)**2))/len(Y_test)
    return TrainError,TestError


##################################################
#Repeat 3a with online training
##################################################
polynomial_tr_error = []
polynomial_te_error = []
gaussian_tr_error = []
gaussian_te_error = []
sigmoid_tr_error = []
sigmoid_te_error = []
M_range = list(range(6))
for M in M_range:
    w_opt_po = OnlineTraining(x_train_1,t_train_1,x_test_1, t_test_1, 'polynomial',M,Epochs=1000,BatchSize=20,stepSize=0.01)
    w_opt_ga = OnlineTraining(x_train_1,t_train_1,x_test_1, t_test_1, 'gaussian',M,Epochs=1000,BatchSize=20,stepSize=0.01)
    w_opt_si = OnlineTraining(x_train_1,t_train_1,x_test_1, t_test_1, 'sigmoidal',M,Epochs=1000,BatchSize=20,stepSize=0.01)


    TrainPred_po , TestPred_po = OnlinePred(w_opt_po,x_train_1,x_test_1,'polynomial')
    TrainPred_ga , TestPred_ga = OnlinePred(w_opt_ga,x_train_1,x_test_1,'gaussian')
    TrainPred_si , TestPred_si = OnlinePred(w_opt_si,x_train_1,x_test_1,'sigmoidal')

    TrainError_po, TestError_po = OnlinePred_Error(w_opt_po,x_train_1,t_train_1,x_test_1,t_test_1,'polynomial')
    TrainError_ga, TestError_ga = OnlinePred_Error(w_opt_ga,x_train_1,t_train_1,x_test_1,t_test_1,'gaussian')
    TrainError_si, TestError_si = OnlinePred_Error(w_opt_si,x_train_1,t_train_1,x_test_1,t_test_1,'sigmoidal')

    #Plot the predicted training and testing targets alongside the original targets for various model orders and all three different kernels.
    fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(15, 15))
    fig.suptitle(f'M = {M}', fontweight='bold')

    fig.text(0.5, 0.96, 'Polynomial Kernel ', ha='center', fontsize=10, fontweight='bold')
    ax1.set_title("Training Estimates vs Original Training Targets")
    ax1.scatter(x_train_1,t_train_1, color='r')
    ax1.scatter(x_train_1,TrainPred_po, color='g')
    ax1.set_xlabel('Training Data')
    ax1.set_ylabel('Training Targets')
    ax1.legend(["Original","Predicted"])
    ax2.set_title("Test Estimates vs Original Test Targets")
    ax2.scatter(x_test_1,t_test_1, color='r')
    ax2.scatter(x_test_1,TestPred_po, color='g')
    ax2.set_xlabel('Testing Data')
    ax2.set_ylabel('Testing Targets')
    ax2.legend(["Original","Predicted"])

    fig.text(0.5, 0.64, 'Gaussian Kernel ', ha='center', fontsize=10, fontweight='bold')
    ax3.set_title("Training Estimates vs Original Training Targets")
    ax3.scatter(x_train_1,t_train_1, color='r')
    ax3.scatter(x_train_1,TrainPred_ga, color='g')
    ax3.set_xlabel('Training Data')
    ax3.set_ylabel('Training Targets')
    ax3.legend(["Original","Predicted"])
    ax4.set_title("Test Estimates vs Original Test Targets")
    ax4.scatter(x_test_1,t_test_1, color='r')
    ax4.scatter(x_test_1,TestPred_ga, color='g')
    ax4.set_xlabel('Testing Data')
    ax4.set_ylabel('Testing Targets')
    ax4.legend(["Original","Predicted"])

    fig.text(0.5, 0.32, 'Sigmoidal Kernel ', ha='center', fontsize=10, fontweight='bold')
    ax5.set_title("Training Estimates vs Original Training Targets")
    ax5.scatter(x_train_1,t_train_1, color='r')
    ax5.scatter(x_train_1,TrainPred_si, color='g')
    ax5.set_xlabel('Training Data')
    ax5.set_ylabel('Training Targets')
    ax5.legend(["Original","Predicted"])
    ax6.set_title("Test Estimates vs Original Test Targets")
    ax6.scatter(x_test_1,t_test_1, color='r')
    ax6.scatter(x_test_1,TestPred_si, color='g')
    ax6.set_xlabel('Testing Data')
    ax6.set_ylabel('Testing Targets')
    ax6.legend(["Original","Predicted"])

    plt.tight_layout()
    plt.show()

    #Store them for plotting
    polynomial_tr_error.append(TrainError_po)
    polynomial_te_error.append(TestError_po)
    gaussian_tr_error.append(TrainError_ga)
    gaussian_te_error.append(TestError_ga)
    sigmoid_tr_error.append(TrainError_si)
    sigmoid_te_error.append(TestError_si)

polynomial_tr_error = np.array(polynomial_tr_error)
polynomial_te_error = np.array(polynomial_te_error)
gaussian_tr_error = np.array(gaussian_tr_error)
gaussian_te_error = np.array(gaussian_te_error)
sigmoid_tr_error = np.array(sigmoid_tr_error)
sigmoid_te_error = np.array(sigmoid_te_error)

# #Plot training error vs polynomial order and testing error vs polynomial order for all the three different kernels

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('Training and Testing Errors vs Polynomial Order', fontweight='bold')

fig.text(0.5, 0.96, 'Polynomial Kernel ', ha='center', fontsize=10, fontweight='bold')
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,polynomial_tr_error, color='r')
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training MSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,polynomial_te_error, color='r')
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing MSE')

fig.text(0.5, 0.64, 'Gaussian Kernel ', ha='center', fontsize=10, fontweight='bold')
ax3.set_title("Training Error vs Polynomial Order")
ax3.plot(M_range,gaussian_tr_error, color='r')
ax3.set_xlabel('Polynomial Order')
ax3.set_ylabel('Training MSE')
ax4.set_title("Test Error vs Polynomial Order")
ax4.plot(M_range,gaussian_te_error, color='r')
ax4.set_xlabel('Polynomial Order')
ax4.set_ylabel('Testing MSE')

fig.text(0.5, 0.32, 'Sigmoidal Kernel ', ha='center', fontsize=10, fontweight='bold')
ax5.set_title("Training Error vs Polynomial Order")
ax5.plot(M_range,sigmoid_tr_error, color='r')
ax5.set_xlabel('Polynomial Order')
ax5.set_ylabel('Training MSE')
ax6.set_title("Test Error vs Polynomial Order")
ax6.plot(M_range,sigmoid_te_error, color='r')
ax6.set_xlabel('Polynomial Order')
ax6.set_ylabel('Testing MSE')

plt.tight_layout()
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10))
fig.suptitle('Training and Testing Errors vs Polynomial Order', fontweight='bold')
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,polynomial_tr_error, color='red')
ax1.plot(M_range,gaussian_tr_error, color='blue')
ax1.plot(M_range,sigmoid_tr_error, color='green')
ax1.legend(["Polynomial","Gaussian","Sigmoid"])
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training MSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,polynomial_te_error, color='red')
ax2.plot(M_range,gaussian_te_error, color='blue')
ax2.plot(M_range,sigmoid_te_error, color='green')
ax2.legend(["Polynomial","Gaussian","Sigmoid"])
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing MSE')

plt.tight_layout()
plt.show()

##################################################
#Repeat 3b with online training
##################################################

polynomial_tr_error = []
polynomial_te_error = []
gaussian_tr_error = []
gaussian_te_error = []
sigmoid_tr_error = []
sigmoid_te_error = []
M_range = range(6)
for M in M_range:
    w_opt_po = OnlineTraining(x_train_2,t_train_2,x_test_2, t_test_2, 'polynomial',M,Epochs=1000,BatchSize=20,stepSize=0.01)
    w_opt_ga = OnlineTraining(x_train_2,t_train_2,x_test_2, t_test_2, 'gaussian',M,Epochs=1000,BatchSize=20,stepSize=0.01)
    w_opt_si = OnlineTraining(x_train_2,t_train_2,x_test_2, t_test_2, 'sigmoidal',M,Epochs=1000,BatchSize=20,stepSize=0.01)


    TrainPred_po , TestPred_po = OnlinePred(w_opt_po,x_train_2,x_test_2,'polynomial')
    TrainPred_ga , TestPred_ga = OnlinePred(w_opt_ga,x_train_2,x_test_2,'gaussian')
    TrainPred_si , TestPred_si = OnlinePred(w_opt_si,x_train_2,x_test_2,'sigmoidal')

    TrainError_po, TestError_po = OnlinePred_Error(w_opt_po,x_train_2,t_train_2,x_test_2,t_test_2,'polynomial')
    TrainError_ga, TestError_ga = OnlinePred_Error(w_opt_ga,x_train_2,t_train_2,x_test_2,t_test_2,'gaussian')
    TrainError_si, TestError_si = OnlinePred_Error(w_opt_si,x_train_2,t_train_2,x_test_2,t_test_2,'sigmoidal')

    #Plot the predicted training and testing targets alongside the original targets for various model orders and all three different kernels.
    fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(15, 15))
    fig.suptitle(f'M = {M}', fontweight='bold')

    fig.text(0.5, 0.96, 'Polynomial Kernel ', ha='center', fontsize=10, fontweight='bold')
    ax1.set_title("Training Estimates vs Original Training Targets")
    ax1.scatter(x_train_2,t_train_2, color='r')
    ax1.scatter(x_train_2,TrainPred_po, color='g')
    ax1.set_xlabel('Training Data')
    ax1.set_ylabel('Training Targets')
    ax1.legend(["Original","Predicted"])
    ax2.set_title("Test Estimates vs Original Test Targets")
    ax2.scatter(x_test_2,t_test_2, color='r')
    ax2.scatter(x_test_2,TestPred_po, color='g')
    ax2.set_xlabel('Testing Data')
    ax2.set_ylabel('Testing Targets')
    ax2.legend(["Original","Predicted"])

    fig.text(0.5, 0.64, 'Gaussian Kernel ', ha='center', fontsize=10, fontweight='bold')
    ax3.set_title("Training Estimates vs Original Training Targets")
    ax3.scatter(x_train_2,t_train_2, color='r')
    ax3.scatter(x_train_2,TrainPred_ga, color='g')
    ax3.set_xlabel('Training Data')
    ax3.set_ylabel('Training Targets')
    ax3.legend(["Original","Predicted"])
    ax4.set_title("Test Estimates vs Original Test Targets")
    ax4.scatter(x_test_2,t_test_2, color='r')
    ax4.scatter(x_test_2,TestPred_ga, color='g')
    ax4.set_xlabel('Testing Data')
    ax4.set_ylabel('Testing Targets')
    ax4.legend(["Original","Predicted"])

    fig.text(0.5, 0.32, 'Sigmoidal Kernel ', ha='center', fontsize=10, fontweight='bold')
    ax5.set_title("Training Estimates vs Original Training Targets")
    ax5.scatter(x_train_2,t_train_2, color='r')
    ax5.scatter(x_train_2,TrainPred_si, color='g')
    ax5.set_xlabel('Training Data')
    ax5.set_ylabel('Training Targets')
    ax5.legend(["Original","Predicted"])
    ax6.set_title("Test Estimates vs Original Test Targets")
    ax6.scatter(x_test_2,t_test_2, color='r')
    ax6.scatter(x_test_2,TestPred_si, color='g')
    ax6.set_xlabel('Testing Data')
    ax6.set_ylabel('Testing Targets')
    ax6.legend(["Original","Predicted"])

    plt.tight_layout()
    plt.show()

    #Store them for plotting
    polynomial_tr_error.append(TrainError_po)
    polynomial_te_error.append(TestError_po)
    gaussian_tr_error.append(TrainError_ga)
    gaussian_te_error.append(TestError_ga)
    sigmoid_tr_error.append(TrainError_si)
    sigmoid_te_error.append(TestError_si)

polynomial_tr_error = np.array(polynomial_tr_error)
polynomial_te_error = np.array(polynomial_te_error)
gaussian_tr_error = np.array(gaussian_tr_error)
gaussian_te_error = np.array(gaussian_te_error)
sigmoid_tr_error = np.array(sigmoid_tr_error)
sigmoid_te_error = np.array(sigmoid_te_error)

# #Plot training error vs polynomial order and testing error vs polynomial order for all the three different kernels

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('Training and Testing Errors vs Polynomial Order', fontweight='bold')

fig.text(0.5, 0.96, 'Polynomial Kernel ', ha='center', fontsize=10, fontweight='bold')
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,polynomial_tr_error, color='r')
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training MSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,polynomial_te_error, color='r')
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing MSE')

fig.text(0.5, 0.64, 'Gaussian Kernel ', ha='center', fontsize=10, fontweight='bold')
ax3.set_title("Training Error vs Polynomial Order")
ax3.plot(M_range,gaussian_tr_error, color='r')
ax3.set_xlabel('Polynomial Order')
ax3.set_ylabel('Training MSE')
ax4.set_title("Test Error vs Polynomial Order")
ax4.plot(M_range,gaussian_te_error, color='r')
ax4.set_xlabel('Polynomial Order')
ax4.set_ylabel('Testing MSE')

fig.text(0.5, 0.32, 'Sigmoidal Kernel ', ha='center', fontsize=10, fontweight='bold')
ax5.set_title("Training Error vs Polynomial Order")
ax5.plot(M_range,sigmoid_tr_error, color='r')
ax5.set_xlabel('Polynomial Order')
ax5.set_ylabel('Training MSE')
ax6.set_title("Test Error vs Polynomial Order")
ax6.plot(M_range,sigmoid_te_error, color='r')
ax6.set_xlabel('Polynomial Order')
ax6.set_ylabel('Testing MSE')

plt.tight_layout()
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10))
fig.suptitle('Training and Testing Errors vs Polynomial Order', fontweight='bold')
ax1.set_title("Training Error vs Polynomial Order")
ax1.plot(M_range,polynomial_tr_error, color='red')
ax1.plot(M_range,gaussian_tr_error, color='blue')
ax1.plot(M_range,sigmoid_tr_error, color='green')
ax1.legend(["Polynomial","Gaussian","Sigmoid"])
ax1.set_xlabel('Polynomial Order')
ax1.set_ylabel('Training MSE')
ax2.set_title("Test Error vs Polynomial Order")
ax2.plot(M_range,polynomial_te_error, color='red')
ax2.plot(M_range,gaussian_te_error, color='blue')
ax2.plot(M_range,sigmoid_te_error, color='green')
ax2.legend(["Polynomial","Gaussian","Sigmoid"])
ax2.set_xlabel('Polynomial Order')
ax2.set_ylabel('Testing MSE')

plt.tight_layout()
plt.show()


########################################
# Study the effect of stepSize on the convergence of weights (plot required results)
########################################

# Define the range of step sizes to study
stepSize = [1e-7, 1e-5, 1e-3, 0.1, 10]

# Lists to store training and testing errors for each kernel
polynomial_train_errors = []
polynomial_test_errors = []
gaussian_train_errors = []
gaussian_test_errors = []
sigmoid_train_errors = []
sigmoid_test_errors = []

# Define epochs and batch size
epochs = 200
batch_size = 10

for step_size in stepSize:
    # Perform online training for each kernel with the current step size
    w_opt_polynomial = OnlineTraining(x_train_2, t_train_2, x_test_2, t_test_2, 'polynomial', 7, epochs, batch_size, step_size)
    w_opt_gaussian = OnlineTraining(x_train_2, t_train_2, x_test_2, t_test_2, 'gaussian', 7, epochs, batch_size, step_size)
    w_opt_sigmoid = OnlineTraining(x_train_2, t_train_2, x_test_2, t_test_2, 'sigmoidal', 7, epochs, batch_size, step_size)

    # Compute training and testing errors for each kernel
    polynomial_tr_error, polynomial_te_error = ErrorPred(w_opt_polynomial, x_train_2, t_train_2, x_test_2, t_test_2, 'polynomial')
    gaussian_tr_error, gaussian_te_error = ErrorPred(w_opt_gaussian, x_train_2, t_train_2, x_test_2, t_test_2, 'gaussian')
    sigmoid_tr_error, sigmoid_te_error = ErrorPred(w_opt_sigmoid, x_train_2, t_train_2, x_test_2, t_test_2, 'sigmoidal')

    # Append errors to respective lists
    polynomial_train_errors.append(polynomial_tr_error)
    polynomial_test_errors.append(polynomial_te_error)
    gaussian_train_errors.append(gaussian_tr_error)
    gaussian_test_errors.append(gaussian_te_error)
    sigmoid_train_errors.append(sigmoid_tr_error)
    sigmoid_test_errors.append(sigmoid_te_error)

# Plot the results for the effect of step size
fig, ax = plt.subplots(2, 3, figsize=(18, 10))

ax[0, 0].plot(stepSize, polynomial_train_errors, label='Training Error')
ax[0, 0].plot(stepSize, polynomial_test_errors, label='Testing Error')
ax[0, 0].set_title('Polynomial Kernel')
ax[0, 0].set_xlabel('Step Size')
ax[0, 0].set_ylabel('Error')
ax[0, 0].legend()

ax[0, 1].plot(stepSize, gaussian_train_errors, label='Training Error')
ax[0, 1].plot(stepSize, gaussian_test_errors, label='Testing Error')
ax[0, 1].set_title('Gaussian Kernel')
ax[0, 1].set_xlabel('Step Size')
ax[0, 1].set_ylabel('Error')
ax[0, 1].legend()

ax[0, 2].plot(stepSize, sigmoid_train_errors, label='Training Error')
ax[0, 2].plot(stepSize, sigmoid_test_errors, label='Testing Error')
ax[0, 2].set_title('Sigmoidal Kernel')
ax[0, 2].set_xlabel('Step Size')
ax[0, 2].set_ylabel('Error')
ax[0, 2].legend()

plt.tight_layout()
plt.show()

########################################
# Study the effect of batch size on the speed of convergence (plot required results)
########################################

# Define the range of batch sizes to study
batch_sizes = [5, 10, 20, 25]

# Lists to store training and testing errors for each kernel
polynomial_train_errors = []
polynomial_test_errors = []
gaussian_train_errors = []
gaussian_test_errors = []
sigmoid_train_errors = []
sigmoid_test_errors = []

# Define epochs and step size
epochs = 200
step_size = 0.01

for batch_size in batch_sizes:
    # Perform online training for each kernel with the current batch size
    w_opt_polynomial = OnlineTraining(x_train_2, t_train_2, x_test_2, t_test_2, 'polynomial', 7, epochs, batch_size, step_size)
    w_opt_gaussian = OnlineTraining(x_train_2, t_train_2, x_test_2, t_test_2, 'gaussian', 7, epochs, batch_size, step_size)
    w_opt_sigmoid = OnlineTraining(x_train_2, t_train_2, x_test_2, t_test_2, 'sigmoidal', 7, epochs, batch_size, step_size)

    # Compute training and testing errors for each kernel
    polynomial_tr_error, polynomial_te_error = ErrorPred(w_opt_polynomial, x_train_2, t_train_2, x_test_2, t_test_2, 'polynomial')
    gaussian_tr_error, gaussian_te_error = ErrorPred(w_opt_gaussian, x_train_2, t_train_2, x_test_2, t_test_2, 'gaussian')
    sigmoid_tr_error, sigmoid_te_error = ErrorPred(w_opt_sigmoid, x_train_2, t_train_2, x_test_2, t_test_2, 'sigmoidal')

    # Append errors to respective lists
    polynomial_train_errors.append(polynomial_tr_error)
    polynomial_test_errors.append(polynomial_te_error)
    gaussian_train_errors.append(gaussian_tr_error)
    gaussian_test_errors.append(gaussian_te_error)
    sigmoid_train_errors.append(sigmoid_tr_error)
    sigmoid_test_errors.append(sigmoid_te_error)

# Plot the results for the effect of batch size
fig, ax = plt.subplots(2, 3, figsize=(18, 10))

ax[0, 0].plot(batch_sizes, polynomial_train_errors, label='Training Error')
ax[0, 0].plot(batch_sizes, polynomial_test_errors, label='Testing Error')
ax[0, 0].set_title('Polynomial Kernel')
ax[0, 0].set_xlabel('Batch Size')
ax[0, 0].set_ylabel('Error')
ax[0, 0].legend()

ax[0, 1].plot(batch_sizes, gaussian_train_errors, label='Training Error')
ax[0, 1].plot(batch_sizes, gaussian_test_errors, label='Testing Error')
ax[0, 1].set_title('Gaussian Kernel')
ax[0, 1].set_xlabel('Batch Size')
ax[0, 1].set_ylabel('Error')
ax[0, 1].legend()

ax[0, 2].plot(batch_sizes, sigmoid_train_errors, label='Training Error')
ax[0, 2].plot(batch_sizes, sigmoid_test_errors, label='Testing Error')
ax[0, 2].set_title('Sigmoidal Kernel')
ax[0, 2].set_xlabel('Batch Size')
ax[0, 2].set_ylabel('Error')
ax[0, 2].legend()

plt.tight_layout()
plt.show()


```

```python
#Real world problems
#Part(2): Breast Cancer Wisconsin (Prognostic) Data Set

from google.colab import drive
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

drive.mount('/content/drive',force_remount=True)
#Download the data from the given link and keep it in the following location
datapath = "/content/drive/MyDrive/PRML_Assignments/Assignment_2/BreastCancer/wpbc.data"
data = []
with open(datapath) as fp:
    Lines = fp.readlines()
    for line in Lines:
      data.append(line.strip().split(',')[2:33])
data = np.array(data)
data = data.astype('float32')

X = (data - data.mean(axis=0)) / data.std(axis=0)
X_train = np.hstack((np.ones((X.shape[0], 1)), X))

y = data[:, 0]

y = y.astype('float64')

def errorFunction(X_train, y, w_opt):
    Pred = X_train.dot(w_opt)
    return (1 / (2 * len(y))) * np.sum((Pred - y)**2)

def gradient_descent(X_train, y, w_opt, alpha, iterations):
    J = np.zeros(iterations)
    for i in range(iterations):
        Pred = X_train.dot(w_opt)
        gradient = (1 / len(y)) * X_train.T.dot(Pred - y)
        w_opt -= alpha * gradient
        J[i] = errorFunction(X_train, y, w_opt)
    return w_opt, J

alpha = 0.01
iterations = 1000

np.random.seed(42)
indices = np.random.permutation(X_train.shape[0])
index = int(0.8 * X_train.shape[0])
X_train, X_test = X_train[indices[:index]], X_train[indices[index:]]
y_train, y_test = y[indices[:index]], y[indices[index:]]

w_opt = np.random.randn(X_train.shape[1])
w_opt, J = gradient_descent(X_train, y_train, w_opt, alpha, iterations)

TrainPred = X_train @ w_opt
TestPred = X_test @ w_opt

# Plotting
plt.scatter(list(range(len(y_train))), y_train)
plt.scatter(list(range(len(TrainPred))), TrainPred, color = 'r')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend(['Actual', 'Predicted'])
plt.title('Actual vs Predicted (Train Set)')
plt.show()
plt.scatter(list(range(len(y_test))), y_test)
plt.scatter(list(range(len(TestPred))), TestPred, color = 'r')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend(['Actual', 'Predicted'])
plt.title('Actual vs Predicted (Test Set)')
plt.show()

# plt.plot(J)
# plt.xlabel('Iterations')
# plt.ylabel('Cost')
# plt.title('Cost History')
# plt.show()

# Evaluate the model
TestError = errorFunction(X_test, y_test, w_opt)
TrainError = errorFunction(X_train, y_train, w_opt)
print("Testing Error :", TestError)
print("Training Error :", TrainError)



```

```python
from scipy.io import arff
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the mounted Google Drive
data_path = "/content/drive/MyDrive/PRML_Assignments/Assignment_2/SupplyChain/"
data, meta = arff.loadarff(data_path + 'scm1d-train.arff')
df = pd.DataFrame(data)

# Separate features and targets
features = df.drop(columns=['timeunit', 'storageCost']).values
targets = df[['timeunit', 'storageCost']].values

# Split data into training and testing sets
train_size = int(0.8 * len(features))
X_train, X_test = features[:train_size], features[train_size:]
y_train, y_test = targets[:train_size], targets[train_size:]

# Add bias term to feature matrices
X_train = np.hstack((np.ones((X_train.shape[0], 1)), X_train))
X_test = np.hstack((np.ones((X_test.shape[0], 1)), X_test))

# Define a function for linear regression with optional regularization
def linear_regression(X, y, lam=0.01):
    identity_matrix = np.eye(X.shape[1])
    # Introduce regularization by adding lam*I
    return np.linalg.inv(X.T @ X + lam * identity_matrix) @ X.T @ y

# Train the model with regularization
weights = linear_regression(X_train, y_train)

# Make predictions
y_pred_train = X_train @ weights
y_pred_test = X_test @ weights

# Define a function to calculate Root Mean Squared Error (RMSE)
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

# Calculate RMSE for training and testing sets
train_rmse = rmse(y_train, y_pred_train)
test_rmse = rmse(y_test, y_pred_test)

# Print RMSE values
print(f"Train RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")
# Visualize the results with enhanced graphs
plt.figure(figsize=(14, 6))

# Plot for Target 1: Actual vs. Predicted
plt.subplot(1, 2, 1)
plt.scatter(y_test[:, 0], y_pred_test[:, 0], color='lightblue', label='Predicted', alpha=0.8, edgecolor='black', linewidth=0.5)
plt.plot([min(y_test[:, 0]), max(y_test[:, 0])], [min(y_test[:, 0]), max(y_test[:, 0])], color='salmon', linestyle='--', label='Ideal', linewidth=2)
plt.xlabel('Actual Target 1')
plt.ylabel('Predicted Target 1')
plt.title('Target 1: Actual vs. Predicted')
plt.legend()
plt.grid(True)
# Plot for Target 2: Actual vs. Predicted
plt.subplot(1, 2, 2)
plt.scatter(y_test[:, 1], y_pred_test[:, 1], color='lightblue', label='Predicted', alpha=0.8, edgecolor='black', linewidth=1.5, marker='o')
plt.plot([min(y_test[:, 1]), max(y_test[:, 1])], [min(y_test[:, 1]), max(y_test[:, 1])], color='red', linestyle='--', label='Ideal', linewidth=2)
plt.xlabel('Actual Target 2')
plt.ylabel('Predicted Target 2')
plt.title('Target 2: Actual vs. Predicted')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()


# Scatter plot with improved visualization for Target 2
plt.figure(figsize=(8, 6))
plt.scatter(y_test[:, 1], y_pred_test[:, 1], color='red', alpha=0.7, edgecolor='black', label='Predicted vs Actual', marker='o')
plt.plot([y_test[:, 1].min(), y_test[:, 1].max()], [y_test[:, 1].min(), y_test[:, 1].max()], 'k--', lw=2, label='Identity Line')
plt.xlabel('Actual Target 2')
plt.ylabel('Predicted Target 2')
plt.title('Target 2: Actual vs. Predicted Scatter Plot with Identity Line')
plt.legend()
plt.grid(True)
plt.show()

```
